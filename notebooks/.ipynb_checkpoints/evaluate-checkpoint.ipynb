```python
# công thức pin ball = MSError= |y-y^| nhân với a
import numpy as np
def pinball_loss(y,y_pred,alpha):
    diff = y- y_pred
    return np.mean(np.maximum(alpha*diff,(alpha-1)*diff)) # maximum không phải max
```
